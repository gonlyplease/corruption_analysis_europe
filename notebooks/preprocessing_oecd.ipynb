```python
import pandas as pd
import numpy as np

oecd_data = pd.read_csv("../data/raw/qogdata_03_01_2025.csv", sep=";")
countries = pd.read_csv("../data/raw/country_mapping.csv")
#print(oecd_data.head(10))
countries.head(100)
oecd_data.head(100)
```

```python
oecd_data.columns
```

```python
filtered_data = oecd_data.merge(countries, how='inner', left_on='cname', right_on='Country')
filtered_data = filtered_data.drop(columns=['Country', 'ISO2 Code', 'ccode', 'ccode_qog', 'cname_qog', 'ccodealp', 'ccodecow', 'version', 'cname_year', 'ccodealp_year'])
filtered_data['year'] = pd.to_datetime(filtered_data['year'], format='%Y', errors='coerce')
filtered_data = filtered_data[
    (filtered_data['year'].dt.year >= 2003) & (filtered_data['year'].dt.year <= 2023)
]

# Renaming cryptic column names
column_mapping = {
    "wjp_cj_cor": "Civil Justice is Free of Corruption",
    "wjp_crsys_cor": "Criminal System is Free of Corruption",
    "ccp_cc": "Corruption Commission Present in Constitution",
    "bti_cr": "Civil Rights",
    "bti_acp": "Anti Corruption Policy",
    "bci_bci": "The Bayesian Corruption Indicator",
}
filtered_data = filtered_data.rename(columns=column_mapping)
filtered_data.head(10000)
```

```python
filtered_data.to_csv("../data/processed/oecd.csv", index=False, index_label=False)
```
