```python
wgi_data = pd.read_excel("../data/raw/wgidataset.xlsx")
```

```python
wgi_data.drop(columns=["codeindyr", "stddev",	"nsource"	,"pctrank",	"pctranklower",	"pctrankupper"], inplace=True)
```

```python
wgi_data.shape
```

```python
wgi_data
wgi_data = wgi_data.pivot(index=["code", "countryname", "year"], columns="indicator", values="estimate").reset_index()
wgi_data.columns.name = None

rename_dict = {
    "cc": "control_of_corruption",
    "ge": "government_effectiveness",
    "pv": "political_stability_and_absence_of_violence_terrorism",
    "rl": "rule_of_law",
    "rq": "regulatory_quality",
    "va": "voice_and_accountability"
}
wgi_data = wgi_data.rename(columns=rename_dict)

wgi_data
```
