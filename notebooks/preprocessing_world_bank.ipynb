```python
import pandas as pd
import numpy as np

world_bank_data = pd.read_csv("../data/raw/world_bank_development_indicators.csv")
countries = pd.read_csv("../data/raw/country_mapping.csv")
print(world_bank_data.head(10))
countries.head()
```

```python
world_bank_data.columns
```

```python
filtered_data = world_bank_data.merge(countries, how='inner', left_on='country', right_on='Country')
filtered_data = filtered_data.drop(columns=['Country', 'ISO2 Code', 'land_area', 'other_greenhouse_emisions', 'real_interest_rate', 'risk_premium_on_lending', 'tax_revenue%', 'goverment_effectiveness_std', 'doing_business', 'time_to_get_operation_license', 'statistical_performance_indicators', 'logistic_performance_index', 'GDP_current_US', 'political_stability_std', 'rule_of_law_std', 'regulatory_quality_std', 'voice_and_accountability_std', 'human_capital_index', 'multidimensional_poverty_headcount_ratio%', 'electric_power_consumption'])
filtered_data['date'] = pd.to_datetime(filtered_data['date'])
filtered_data = filtered_data[
    (filtered_data['date'].dt.year >= 2005) & (filtered_data['date'].dt.year <= 2023)
]
filtered_data.head()
```

```python
filtered_data.to_csv("../data/processed/world_bank.csv", index=False, index_label=False)
```
