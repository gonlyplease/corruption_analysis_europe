```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/complete_data.csv")
```

```python


# For inline plots in Jupyter
%matplotlib inline


# Quick overview of the dataframe
print("DataFrame Shape:", df.shape)
print(df.info())
df.head(5)
```

```python
df['The Bayesian Corruption Indicator'] = df['The Bayesian Corruption Indicator'].str.replace(',', '.').astype(float)
```

```python
# 3. Data Cleaning & Preparation
# 

# A) Check for duplicates
duplicate_rows = df[df.duplicated()]
print("Number of duplicate rows:", len(duplicate_rows))

# B) Drop duplicates (if relevant)
df.drop_duplicates(inplace=True)

# C) Identify missing values
missing_values = df.isna().sum()
print("Missing values in each column:\n", missing_values)

# (Optional) You can decide to drop columns or rows with excessive missing data, 
# or fill/impute them if necessary.
# Example: 
# df = df.dropna(subset=["Corruption Perceptions Index Rank", "year"])  # if those columns are essential

# D) Convert data types if needed
# For example, if 'gdp' is stored as object, convert it to numeric
# (Make sure to handle non-numeric values or cleaning first)
df['gdp'] = pd.to_numeric(df['gdp'], errors='coerce')

# Print a summary after cleaning
df.info()
```

```python
# 5. Focus on Italy

italy_df = df[df['country'] == "Italy"].copy()  # Create a subset for Italy
print("Number of records for Italy:", italy_df.shape[0])

# Look at the range of years available for Italy
print("Italy data years range:", italy_df['year'].min(), "-", italy_df['year'].max())

# Examine columns related to real vs. perceived corruption
# For instance, 'The Bayesian Corruption Indicator' might serve as 'real corruption' measure 
# (if that’s how you define it) or you may define real corruption differently.
# 'Corruption Perceptions Index Score' or 'Perceptions Of Corruption' might represent perceived corruption.

italy_corruption_cols = [
    "year",
    "Corruption Commission Present in Constitution",
    "The Bayesian Corruption Indicator",
    "Corruption Perceptions Index Score",
    "Perceptions Of Corruption",
    "Corruption Perceptions Index Standard Error",
]
italy_df[italy_corruption_cols].head()
```

```python
# 6. Visual Comparison: Real vs. Perceived Corruption in Italy

# Line plot over the years for Italy: 
# - The Bayesian Corruption Indicator (hypothetical "real" corruption)
# - Corruption Perceptions Index Score (perceived corruption)
# - Perceptions Of Corruption (alternative measure of perceived corruption)

plt.figure(figsize=(12, 6))

# Make sure these columns have numeric values or handle missing data 
# If they are strings, convert them to numeric first
italy_df['The Bayesian Corruption Indicator'] = pd.to_numeric(italy_df['The Bayesian Corruption Indicator'], errors='coerce')
italy_df['Corruption Perceptions Index Score'] = pd.to_numeric(italy_df['Corruption Perceptions Index Score'], errors='coerce')
italy_df['Perceptions Of Corruption'] = pd.to_numeric(italy_df['Perceptions Of Corruption'], errors='coerce')

# Sort by year for a proper time-series
italy_df.sort_values(by='year', inplace=True)

plt.plot(italy_df['year'], italy_df['The Bayesian Corruption Indicator'], marker='o', label='Real Corruption (Bayesian Indicator)')
plt.plot(italy_df['year'], italy_df['Corruption Perceptions Index Score'], marker='o', label='Perceived Corruption (CPI Score)')
plt.plot(italy_df['year'], italy_df['Perceptions Of Corruption'], marker='o', label='Perceptions Of Corruption (Survey)')

plt.title("Real vs. Perceived Corruption in Italy Over Time")
plt.xlabel("Year")
plt.ylabel("Indicator Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7. Statistical Analysis: Correlation Between Indicators

# Correlation matrix for Italy’s relevant indicators
corr_columns = [
    "Corruption Commission Present in Constitution",
    "The Bayesian Corruption Indicator",
    "Corruption Perceptions Index Score",
    "Perceptions Of Corruption",
    "Corruption Perceptions Index Standard Error",
]
italy_corr_df = italy_df[corr_columns].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(italy_corr_df, annot=True, cmap="YlGnBu", vmin=-1, vmax=1)
plt.title("Correlation Matrix - Italy: Real vs. Perceived Corruption")
plt.show()

italy_corr_df
```
